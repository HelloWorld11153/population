```python
#내 생일날 그래프

import csv
f = open('seoul.csv')
data = csv.reader(f)
next(data)
high = []
low = []
year=[]
for row in data:
    if row[4]!='' and row[3]!='':
        date = row[0].split('-')
        if date[1] == '08' and date[2] == '13':
            year.append(int(row[0].split('-')[0]))
            high.append(float(row[4]))
            low.append(float(row[3]))

import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')         #맑은 고딕 글꼴
plt.rcParams['axes.unicode_minus'] = False      #마이너스 기호 깨짐 방지
plt.title('내 생일(8월 13일)의 기온 변화 그래프')
plt.plot(year, high, 'red', label = 'high')
plt.plot(year, low, 'blue', label = 'low')
plt.legend()
plt.show()
```

```python
#1월 8월 최고기온 히스토그램

import csv
f=open('seoul.csv')
data=csv.reader(f)
next(data)
jan=[]
aug=[]

for row in data:
    month=row[0].split('-')[1]
    if row[4]!='':
        if month=='01':
            jan.append(float(row[4]))
        if month=='08':
            aug.append(float(row[4]))

import matplotlib.pyplot as plt
plt.hist(jan, bins=200, label="1월")
plt.hist(aug, bins=200, label="8월")
plt.legend()            #범례
plt.show()
```

```python
#1월 8월 최고기온 박스플롯

import csv
f=open('seoul.csv')
data=csv.reader(f)
next(data)
jan=[]
aug=[]

for row in data:
    month=row[0].split('-')[1]
    if row[4]!='':
        if month=='01':
            jan.append(float(row[4]))
        if month=='08':
            aug.append(float(row[4]))

import matplotlib.pyplot as plt
plt.boxplot([jan, aug])
plt.show()
```

```python
#8월 기온 박스플롯

import csv
f=open('seoul.csv')
data=csv.reader(f)
next(data)

day=[]
for i in range(31):
    day.append([])

for row in data:
    if row[4]!='':
        if row[0].split('-')[1]=='08':
            day[int(row[0].split('-')[2])-1].append(float(row[4]))
            
plt.style.use('ggplot')
plt.figure(figsize=(10,5))
plt.boxplot(day, showfliers=False)           #이상값 표시 X

plt.show()
```
