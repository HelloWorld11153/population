```python
import csv

f=open('age.csv')
data=csv.reader(f)
result=[]
name=input('지역 이름 입력 : ')

for row in data:
    if name in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(',','')))
            
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')                    #이걸 해야 네모로 안뜬다
plt.title(name+' 인구 구조')
plt.plot(result)
plt.show()
```

```python
import csv
f=open('gender.csv')
data=csv.reader(f)
m=[]
f=[]
name=input('지역 입력 : ')

for row in data:
    if name in row[0]:
        for i in row[3:104]:
            m.append(int(i.replace(',','')))
        for i in row[106:]:
            f.append(-int(i.replace(',','')))
        break
            
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False                 # 마이너스 네모표시 해결
plt.title(name+'의 인구 성별 분포')
plt.barh(range(101), m, label='남성')                     # 항아리모양으로 만들기
plt.barh(range(101), f, label='여성')
plt.legend()
plt.show()                                                # 수치를 -로 표현 안하려면?
```

```python
#동 이름이 같은 지역(ex)부흥동) 구분하기 (미완성)
import csv
f=open('gender.csv')
data=csv.reader(f)
name=input('지역 입력 : ')
header=next(data)
m=[]
f=[]

for row in data:
    if name in row[0]:
        for i in row[3:104]:
            m.append(int(i))
        for i in row[106:]:
            f.append(-int(i))
            
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus']=False
plt.title(name+'지역의 성별 인구분포')
plt.barh(range(101),m,label='Male')
plt.barh(range(101),f,label='Female')
plt.legend()
plt.show()
```

```python
import csv
f=open('gender.csv')
data=csv.reader(f)
name=input('지역 입력 : ')
size=[]

for row in data:
    if name in row[0]:
        m=0
        f=0
        for i in range(101):
            m+=int(row[i+3])
            f+=int(row[i+106])
size.append(m)
size.append(f)

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.axis('equal')
plt.pie(size, labels=['남', '여'], autopct='%.1f%%', startangle=90)       #label(남여이름붙이기), auto percentage, startangle기본 3시)
plt.title(name+'의 남녀인구비율')
plt.show()
```

```python
import csv
f=open('gender.csv')
data=csv.reader(f)
name=input('지역 입력 : ')
m=[]
f=[]

for row in data:
    if name in row[0]:
        for i in row[3:104]:
            m.append(int(i.replace(',','')))
        for i in row[106:]:
            f.append(int(i.replace(',','')))

import matplotlib.pyplot as plt
plt.plot(m, label="남")
plt.plot(f, label='여')
plt.legend()
plt.show()
```

```python
import csv
f=open('gender.csv')
data=csv.reader(f)
name=input('지역 입력 : ')
result=[]

for row in data:
    if name in row[0]:
        for i in range(3,104):
            result.append(int(row[i])-int(row[i+103]))

import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.style.use('ggplot')
plt.bar(range(101), result)
plt.rcParams['axes.unicode_minus']=False
plt.title(name+'의 남녀인구차이 (-는 여>남)')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
x=[]
y=[]
size=[]
for i in range(100):
    x.append(random.randint(50,100))
    y.append(random.randint(50,100))
    size.append(random.randint(50,100))
plt.scatter(x, y, c=x, s=size, cmap='jet', alpha=0.7)
plt.colorbar()
plt.show()
```
